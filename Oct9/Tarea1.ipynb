a mean of 70 and a sigma=2.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from math import *

mu = 70
sigma = 2


def phi(x):
    a = np.zeros(len(x))
    for i in range(len(x)):
    #'Cumulative distribution function for the standard normal distribution'
        a[i] = (1.0 + erf(i-70 / sqrt(2.0*2*2))) / 2.0
    return a

H_0 = np.arange(0,100,1)

F = phi(H_0)


plt.plot(F)
plt.xlim(20, 30)
plt.show()



```
